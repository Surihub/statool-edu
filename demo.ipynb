# 기계학습 실습하기

```python
from statool import *

데이터 = 데이터불러오기('펭귄')
데이터 = 숫자로변환(데이터)
미리보기(데이터,3)
```

```python
학습특성, 테스트특성, 학습타겟, 테스트타겟 = 데이터분할하기(데이터, '성별')
미리보기(학습특성)
```

```python
# 선형 모델 평가
선형_모델 = 모델_훈련(학습특성, 학습타겟, '선형')
선형_정확도 = 모델_평가(테스트특성, 테스트타겟, 선형_모델)
print(f'선형 모델 정확도: {선형_정확도}')
```

```python
# 이차식 모델 평가
이차식_모델 = 모델_훈련(학습특성, 학습타겟, '이차식')
이차식_정확도 = 모델_평가(테스트특성, 테스트타겟, 이차식_모델)
print(f'이차식 모델 정확도: {이차식_정확도}')
```

```python
# 랜덤 포레스트 모델 평가
랜덤_포레스트_모델 = 모델_훈련(학습특성, 학습타겟, '랜덤 포레스트')
랜덤_포레스트_정확도 = 모델_평가(테스트특성, 테스트타겟, 랜덤_포레스트_모델)
print(f'랜덤 포레스트 모델 정확도: {랜덤_포레스트_정확도}')
```
